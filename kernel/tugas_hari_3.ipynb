```python
nama_produk = input()
jumlah_produk_terjual = int(input())
harga_produk = int(input())
```

```python
total_revenue = jumlah_produk_terjual * harga_produk

def kategori_penjualan(total_revenue):
  if total_revenue <= 2000000:
    print('Rendah')
  elif total_revenue >= 2000000 | total_revenue <= 8000000:
    print('Sedang')
  elif total_revenue > 8000000:
    print('Tinggi')
print('Nama produk:',nama_produk)
print('Jumlah produk:',jumlah_produk_terjual)
print('Harga produk:',harga_produk)
print('Total revenue',total_revenue)
kategori_penjualan(total_revenue)
```

- Bagaimana total revenue dihitung?
jumlah_produk_terjual * harga_produk

- Bagaimana if/else digunakan menentukan kategori penjualan?
di gunakan untuk menentukan kategori jika kurang dari 2000000 'rendah', jika lebih dari 200000 - 8000000 'sedang', dan jika lebih dari 8000000 'tinggi'
